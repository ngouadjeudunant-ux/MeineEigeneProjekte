Die Java API (Application Programming Interface) ist die Standard Klassenbibliothek von Java 

I - Wrapper-Klassen 
    Byte, Short, Integer, Long, Double, Float, Boolean, Character
Wrapper-Objekte kapseln einen primitiven Wert in einem Objket, sodass eine Referenz auf diesem Wert existiert. 
String s1  = "Paul" und String s2 = "Paul" sind das gleiche Okjekt. 


II- Stream 
Ein Stream (dt: Strom) ist eine Datenstruktur, die Informationen sequentiell nach dem FIFO-Prinzip von einer Quelle zu einem Ziel befördert.
Input-Stream und Output-Stream

    Streams Datentypen
    Binärdaten: Daten liegen in einer internen Darstellung vor, z.B Bilder, Tondateien, .exe, .class ... 
        Für Byte Streams gibt es zum Einlesen (InputStream oder eine Unterklasse) und zum Schreiben (OutputStream oder eine Unterklasse)
    Textzeichen : Daten liegen in einem Zeichensatz vor, z.B. .txt-Datei, .java, ... , 
    Für Character Streams gibt es zum Einlesen (Reader oder eine Unterklasse) und zum Schreiben Writer oder eine Unterklasse
